```python
import pandas as pd
```

```python
greek_gods = {
    'God': ['Zeus', 'Poseidon','Hera','Athena','Hades'],
    'Domain': ['King of the Gods', 'God of the Sea', 'Queen of the Gods', 'Goddess of Wisdom', 'God of the Underworld'],
    'Age': [10000, 9000, 8500, 8000,9500],
    'Symbol': ['Thunderbolt', 'Trident', 'Peacock', 'Owl', 'Helmet']
}
```

```python
god_df = pd.DataFrame(greek_gods)
```

```python
god_df
```

```python
greek_goddesses = {
    'Goddess': ['Aphrodite', 'Artemis','Demeter','Hestia','Persephone' ],
    'Domain': ['Goddess of Love', 'Goddess of the Hunt', 'Goddess of the Agriculture', 'Goddess of the Hearth', 'Queen of the Underworld'],
    'Age': [7000, 7500, 8000, 8200,7800],
    'Symbol': ['Dove', 'Bow and Arrow', 'Wheat', 'Fire', 'Pomegranate']
}
```

```python
goddess_df = pd.DataFrame(greek_goddesses)
```

```python
goddess_df
```

```python
#1.Merge the data from greek_gods.csv and greek_goddesses.csv based on a common field and create a new table that includes information about both gods and goddesses.
```

```python
merged_data = pd.merge(god_df, goddess_df , how='outer')
```

```python
merged_data
```

```python
#2.Filter the merged table to only include gods and goddesses who are older than 8000 years, then sort them based on their ages in descending order.
```

```python
filtered_sorted = merged_data[(merged_data['Age'] > 8000)].sort_values(by=['Age'], ascending=False)
```

```python
filtered_sorted
```

```python
#3.Join the two tables based on the "Domain" field and calculate the average age of gods and goddesses in each domain.
```

```python
avg_age = merged_data.groupby('Domain')['Age'].mean()
```

```python
avg_age
```

```python
#4.Determine which god/goddess has the highest age, and then find out if they are a god or a goddess.
```

```python
max_age = merged_data.loc[merged_data['Age'].idxmax()]
```

```python
max_age
```

```python
#5.Create a new column in each table called "Age_Group" and categorize the gods/goddesses into groups such as "Young" (age < 5000), "Middle-aged" (age between 5000 and 8000), and "Old" (age > 8000).
```

```python
bins = [0,5000,8000,float('inf')]
labels = ['Young','Middle-aged','Old']
```

```python
god_df['Age Group'] = pd.cut(god_df['Age'],bins = bins,labels = labels,right = 'False')
```

```python
god_df
```

```python
goddess_df['Age Group'] = pd.cut(goddess_df['Age'],bins = bins,labels = labels,right = 'False')
```

```python
goddess_df
```

```python
#6.Compare the average ages of gods and goddesses. Is there a significant age difference between them? If yes, which group tends to be older?
```

```python
avg_age_god = god_df['Age'].mean()
avg_age_goddess = goddess_df['Age'].mean()
```

```python
if avg_age_god > avg_age_goddess:
    print("Gods tend to be older")
elif avg_age_goddess > avg_age_god:
    print("Goddess tend to be older")
else:
    print("there is no difference between the avg age")
```
